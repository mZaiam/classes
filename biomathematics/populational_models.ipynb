```python
from biomathematics import Models, NumericalMethods

import matplotlib.pyplot as plt
import numpy as np
```

```python
models, numerical_methods = Models(), NumericalMethods()

malthus, verhulst, montroll, gompertz = models.malthus, models.verhulst, models.montroll, models.gompertz
euler, rk4 = numerical_methods.euler, numerical_methods.rk4
```

## Malthus

```python
x0 = 0
y0 = 10
h = 0.01
n = 100
r = 2

x_euler, y_euler = euler(malthus, x0, y0, h, n, r)
x_rk4, y_rk4 = rk4(malthus, x0, y0, h, n, r)

x_analytical = np.linspace(x0, x0 + n * h, n)
y_analytical = y0 * np.exp(r * x_analytical)

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].plot(x_euler, y_euler, label='Euler', linestyle='-', color='blue', marker='o', markersize=4, alpha=0.7)
ax[0].plot(x_rk4, y_rk4, label='RK4', linestyle='-', color='green', marker='s', markersize=4, alpha=0.7)
ax[0].plot(x_analytical, y_analytical, linestyle='--', label='Analytical', color='red', linewidth=2)
ax[0].set_xlabel('Time (x)')
ax[0].set_ylabel('Population (y)')
ax[0].legend()

ax[1].plot(x_euler, y_euler, label='Euler', linestyle='-', color='blue', marker='o', markersize=4, alpha=0.7)
ax[1].plot(x_rk4, y_rk4, label='RK4', linestyle='-', color='green', marker='s', markersize=4, alpha=0.7)
ax[1].plot(x_analytical, y_analytical, linestyle='--', label='Analytical', color='red', linewidth=2)
ax[1].set_xlim(0.8 * x_analytical[-1], x_analytical[-1])
ax[1].set_ylim(0.5 * y_analytical[-1], y_analytical[-1])
ax[1].set_xlabel('Time (x)')
ax[1].set_ylabel('Population (y)')
ax[1].legend()

fig.suptitle('Numerical and Analytical Solutions of the Malthus Model')
plt.show()
```

## Verhulst

```python
x0 = 0
y0 = 20
h = 0.01
n = 500
r = 2
k = 100

x_euler, y_euler = euler(verhulst, x0, y0, h, n, r, k)
x_rk4, y_rk4 = rk4(verhulst, x0, y0, h, n, r, k)

x_analytical = np.linspace(x0, x0 + n * h, n)
y_analytical = k / (1 + (k - y0) / y0 * np.exp(- r * x_analytical))

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].plot(x_euler, y_euler, label='Euler', linestyle='-', color='blue', marker='o', markersize=4, alpha=0.7)
ax[0].plot(x_rk4, y_rk4, label='RK4', linestyle='-', color='green', marker='s', markersize=4, alpha=0.7)
ax[0].plot(x_analytical, y_analytical, linestyle='--', label='Analytical', color='red', linewidth=2)
ax[0].set_xlabel('Time (x)')
ax[0].set_ylabel('Population (y)')
ax[0].legend()

ax[1].plot(x_euler, y_euler, label='Euler', linestyle='-', color='blue', marker='o', markersize=4, alpha=0.7)
ax[1].plot(x_rk4, y_rk4, label='RK4', linestyle='-', color='green', marker='s', markersize=4, alpha=0.7)
ax[1].plot(x_analytical, y_analytical, linestyle='--', label='Analytical', color='red', linewidth=2)
ax[1].set_xlim(x_analytical[100], x_analytical[200])
ax[1].set_ylim(y_analytical[100], y_analytical[200])
ax[1].set_xlabel('Time (x)')
ax[1].set_ylabel('Population (y)')
ax[1].legend()

fig.suptitle('Numerical and Analytical Solutions of the Verhulst Model')
plt.show()
```

## Gompertz

```python
x0 = 0
y0 = 10
h = 0.01
n = 500
k = 50

x_euler, y_euler = euler(gompertz, x0, y0, h, n, k)
x_rk4, y_rk4 = rk4(gompertz, x0, y0, h, n, k)

x_analytical = np.linspace(x0, x0 + n * h, n)
y_analytical = k * np.exp(np.log(y0 / k) * np.exp(- x_analytical))

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].plot(x_euler, y_euler, label='Euler', linestyle='-', color='blue', marker='o', markersize=4, alpha=0.7)
ax[0].plot(x_rk4, y_rk4, label='RK4', linestyle='-', color='green', marker='s', markersize=4, alpha=0.7)
ax[0].plot(x_analytical, y_analytical, linestyle='--', label='Analytical', color='red', linewidth=2)
ax[0].set_xlabel('Time (x)')
ax[0].set_ylabel('Population (y)')
ax[0].legend()

ax[1].plot(x_euler, y_euler, label='Euler', linestyle='-', color='blue', marker='o', markersize=4, alpha=0.7)
ax[1].plot(x_rk4, y_rk4, label='RK4', linestyle='-', color='green', marker='s', markersize=4, alpha=0.7)
ax[1].plot(x_analytical, y_analytical, linestyle='--', label='Analytical', color='red', linewidth=2)
ax[1].set_xlim(x_analytical[100], x_analytical[200])
ax[1].set_ylim(y_analytical[100], y_analytical[200])
ax[1].set_xlabel('Time (x)')
ax[1].set_ylabel('Population (y)')
ax[1].legend()

fig.suptitle('Numerical and Analytical Solutions of the Gompertz Model')
plt.show()
```

## Montroll

```python
x0 = 0
y0 = 20
h = 0.001
n = 5000
r = 2
k = 100
b = [0.1, 1, 10]

for bi in b:
    x_rk4, y_rk4 = rk4(montroll, x0, y0, h, n, r, k, bi)
    plt.plot(x_rk4, y_rk4, label=r'$b=$' + f'{bi}')
    
plt.xlabel('Time (x)')
plt.ylabel('Population (y)')
plt.legend()
plt.title('Numerical Solutions via RK4 of the Montroll Model')
plt.show()
```
